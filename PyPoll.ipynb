```python
#import dependencies 
import os 
import csv
import random
import string
import pandas as pd
```

```python
# Path to collect data from the Resources folder
csvpath = os.path.join('..', 'Resources', 'election_data.csv')
outputpath = os.path.join('..', 'Resources', 'election_results.txt')
```

```python
# Read in the CSV file
with open(csvpath, newline='') as csvfile:
    csvreader = csv.reader(csvfile, delimiter=',')
    print(csvreader)
    csv_header = next(csvreader)
    print(f"CSV Header: {csv_header}")
    for row in csvreader: 
        print(row)
```

```python
# Read in the CSV file
csvpathsum = pd.read_csv(csvpath)
```

```python
#Create a DataFrame with data from CSV
df = pd.DataFrame(csvpathsum)
```

```python
#Look at the top 5 lines of the data to see what it contains 
df.head()
```

```python
#Calculate the total # of votes casted
totalcount = df["Voter ID"].count()
totalcount
```

```python
#Calculate total # of votes each candidate received
candidatevotes = df["Candidate"].value_counts()
candidatevotes
```

```python
#Calculate total % of votes each candidate received
percentagevotes = (df["Candidate"].value_counts() / totalcount)
percent = percentagevotes * 100
percent
```

```python
#Print all election results below
output = ("\nElection Results"
    "\n----------------------\n"
    f"Total Votes: {totalcount}\n"
    "----------------------\n"
    f"Khan: {percent.values[0]}% ({candidatevotes[0]})\n"
    f"Correy: {percent.values[1]}% ({candidatevotes[1]})\n"
    f"Li: {percent.values[2]}% ({candidatevotes[2]})\n"
    f"O'Tooley {percent.values[3]}% ({candidatevotes[3]})\n"
    "----------------------\n"
    f"Winner: {percent.index[0]}\n"
    "----------------------\n")
output
```

```python
with open(outputpath, "w", newline='') as csvfile:
    csvfile.write(output)
```
